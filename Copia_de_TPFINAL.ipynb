# **Ejercicio 2**

Esta es una base de datos obtenida de la página de la Organización de las Naciones Unidas para la Agricultura y la Alimentación (FAO). En las que se analiza el presupuesto de nutrientes especificamente en las áreas de America del Sur y la Union Europea teniendo en cuenta distintos factores de estudio.

El presupuesto de nutrientes de las tierras de cultivo se refiere al equilibrio entre la cantidad de nutrientes que se agregan a un campo o suelo a través de la fertilización y otros insumos, y la cantidad de nutrientes que se retiran del suelo debido a la cosecha de cultivos y otros procesos agrícolas. Es una herramienta esencial en la gestión agrícola para garantizar que los cultivos reciban los nutrientes necesarios para un crecimiento saludable y permite optimizar la cantidad y el tipo de fertilizantes a aplicar para mantener la fertilidad del suelo, minimizar el desperdicio de nutrientes y promover la sostenibilidad agrícola a largo plazo.

Las variables que analiza este set de datos son el Area geográfica analizada, el elemento o macronutriente analizado, el producto (que se refiere a los procesos de nutrición o a los de perdida de nutrientes que suceden en los campos de cultivos), el año analizado, la unidad de medida y los valores numericos.

Ademas del "valor" numerico, a partir del cual se puede seguir la variacion a medida que pasan de los años, otra variable muy relevante es el "producto" que nos indica la accion mediante la cual incrementa o disminuye nuestra concentracion de nutrientes escenciales en el suelo de cultivo.

**Ámbito:**
  Tipo: Categórica.
  Justificación: El "Ámbito" simplemente refiere a la tématica que analiza el dataset, por eso en este caso es el mismo para todos los datos "presupuesto de nutrientes de las tierras de cultivo"

**Área:**
  Tipo: Categórica.
  Justificación: La columna "Área" se refiere a la ubicación geográfica o espacial a la que pertenecen los datos de presupuestos de nutrientes. En este caso regiones.

**Elemento:**
  Tipo: Categórica.
  Justificación: El "Elemento" describe el tipo de información específica sobre los presupuestos de nutrientes. En este caso información sobre los nutrientes individuales.

**Producto:**
  Tipo: Categórica.
  Justificación: La columna "Producto" se refiere a los productos agrícolas específicos relacionados con los presupuestos de nutrientes. Cada entrada representa un metodo de ingreso o salida de dicho nitriente.

**Año:**
  Tipo: Temporal (Fecha o Año).
  Justificación: La columna "Año" indica el año en que se registraron los datos del presupuesto de nutrientes. Esta variable representa información temporal.

**Unidad:**
  Tipo: Categórica.
  Justificación: La "Unidad" se refiere a las unidades de medida utilizadas para cuantificar los valores. En este caso Kg/ha

**Valor:**
  Tipo: Numérica (Continua)
  Justificación: La columna "Valor" contiene datos numéricos que representan cantidades relacionadas con los presupuestos de nutrientes en la agricultura. En este caso vendría a ser el la cantidad de kg del nutriente por cada hectaria de suelo de cultivo.
